```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import codecs 
```

```python
#csv to utf-8 
df = pd.read_csv(r'C:\Users\harr1\Desktop\Thinkful\WELLCOME_APCspend2013_forThinkful.csv', encoding = "ISO-8859-1")
print (df.head(10))
```

```python
df['Article title'].describe()
#Duplicate article entry
```

```python
 
# sorting by article title
df.sort_values("Article title", inplace = True) 
  
# dropping ALL duplicte values 
df.drop_duplicates(subset ="Article title", 
                     keep = 'first', inplace = True) 
  
# displaying data 
df
```

```python
#determine the five most common journals and the total articles for each.
df['Journal title'].value_counts()
```

```python
#Eliminating case sensitivity
sm_titles = df['Journal title'].str.lower()
df['Titles'] = sm_titles
```

```python
#determine the five most common journals and the total articles for each.
sm_titles.value_counts()
```

```python
cost_translate = df['COST (£) charged to Wellcome (inc VAT when charged)'].str.replace('£', " " )
cost_translate_two = cost_translate.str.replace('$', " ")
cost_trans_int = pd.to_numeric(cost_translate_two)
df['cost'] = cost_trans_int
print(df)

```

```python
#Next, calculate the mean, median, and standard deviation of the open-access cost per article for each journal . 

grouped = df.groupby('Titles')
print (grouped['cost'].agg(np.mean))

```

```python
grouped = df.groupby('Titles')
print (grouped['cost'].agg(np.median))
```

```python
grouped = df.groupby('Titles')
print (grouped['cost'].agg(np.std))
```
